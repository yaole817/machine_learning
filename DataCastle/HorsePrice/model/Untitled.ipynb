```python
y = data['price']
```

```python
print(y)
```

```python
xVars = headers = ['date','bedroomNum','bathroomNum','horseAres','carArea','floor','wholeArea',
         'structureArea','basementAres','structureYear','repairYear','col','row']
X = data[xVars]
```

```python
print(X)
```

```python
from sklearn.cross_validation import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
xTrain,xTest,yTrain,yTest = train_test_split(X, y, random_state = 1)
```

```python
print(xTrain.shape)
```

```python
print(xTest.shape)
```

```python
linreg = LinearRegression()
```

```python
model = linreg.fit(xTrain,yTrain)
```

```python
print(model)
```

```python
print(linreg.intercept_)
```

```python
print(linreg.coef_)
```

```python
yPred = linreg.predict(xTest)
```

```python
plt.figure()
```

```python
plt.plot(range(len(yPred)),yPred,'b',label="predict")  
plt.plot(range(len(yTest)),yTest,'r',label="test") 
```
